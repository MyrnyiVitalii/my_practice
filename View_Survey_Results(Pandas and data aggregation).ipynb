## 5. Який відсоток респондентів програмує на Python  
Процент людей указавших, что программируют на Python и других языках: **48.39%**  
Процент людей программирующих исключительно на Python: **1.3%**  

```python
survey['LanguageHaveWorkedWith'] = survey['LanguageHaveWorkedWith'].str.lower()
low  = survey['LanguageHaveWorkedWith'].dropna()
low
```

```python
only_python = survey[survey['LanguageHaveWorkedWith'] == "python"].shape[0]
only_python
# Кол-во у которых только пайтон
```

```python
python_users = survey[survey['LanguageHaveWorkedWith'].str.contains('python', na=False)]
python_users.shape[0]
# Кол-во где пайтон встречался
```

```python
low
other = 0
count_python = 0

for i in low:
    if 'python' in i:
        count_python += 1
    else:
        other += 1
        
total = count_python + other  # Колво людей написавших информацию
percent = (count_python / response_count)*100 # процент от общего количества
percent_r = round(percent,2) # Процент из тех кто указал пайто и другие
percent_only_python = round((only_python/total)*100,2) # процент из тех кто только пайтон

print(f'Количество людей в опросе: {response_count} чел.\nКоличество людей указавщие данные о языке программирования: {total} чел.\nКоличество людей которые упамянули Python: {count_python} чел.\nКоличество людей указавших исключительно Python: {only_python} чел.\n')
print(f'Процент людей указавших, что программируют на Python и других языках:: {percent_r}%')
print(f'Процент людей программирующих исключительно на Python: {percent_only_python}%')
```
